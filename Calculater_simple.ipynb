```python

def calculater():
    num1 = float(input("enter the first number ="))
    operater = input("enter the oprater(+,-,*,/)=")
    num2 = float(input("enter the second number="))

    if operater == "+":
        print(num1 + num2)
    elif operater == "-":
        print(num1 - num2)
    elif operater == "*":
        print(num1 * num2)
    elif operater == "/":
        if num2 == 0:
            print("not defined")
        else:
            print(num1 / num2)
    else:
        print("invalid operator")

calculater()
```
