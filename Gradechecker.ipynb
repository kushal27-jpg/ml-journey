```python

students = {
    "Kael": 85,
    "Meha": 92,
    "Veer": 67,
    "Arjun": 78,
    "Priya": 55
}

for name, score in students.items():
    if score >= 80:
        print(name, "- Pass with distinction")
    elif score >= 60:
        print(name, "- Pass")
    else:
        print(name, "- Fail")
```
