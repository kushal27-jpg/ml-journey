```python

task = []

while True:
    print("1.add task")
    print("2.view task")
    print("3.remove task")
    print("4.quit")

    choice = input("enter your choice:")

    if choice == "1":
        goal = input("enter your task :")
        task.append(goal)

    elif choice == "3":
        pop = int(input("enter the number of task to remove "))
        task.pop(pop - 1)

    elif choice == "2":
        for i, t in enumerate(task):
            print(i+1, "-", t)

    elif choice == "4":
        break
```
