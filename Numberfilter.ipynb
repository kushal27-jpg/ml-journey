```python

numbers = []

count = int(input("How many numbers do you want to enter? "))

for i in range(count):
    num = int(input("Enter number: "))
    numbers.append(num)

result = []

for num in numbers:
    if num > 10:
        result.append(num)

print("Numbers greater than 10:", result)
```
